```python
# solution uses:
#   algorithm design
#   functions
#   user input
#   user input checks
#   control flow (if elif else)
#   special operators
#   equality operators
#   logical operators

# input: day of week
# the days of the week are: 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
# output: work status
# work status can be either: 'Holiday' or 'Workday'

def work_status(day_of_week, weekdays):
    status = None
    
    if (day_of_week == 'Saturday') or (day_of_week == 'Sunday'):  # if its the weekend, we have a 'Holiday'
        status = 'Holiday'
    elif day_of_week in weekdays:  # if not the weekend, must be a work day
        status = 'Workday'
    else:  # checks for valid input
        status = 'Invalid Entry'
    
    return status


def our_app():
    # valid user input (also required data)
    weekdays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    
    # force valid user input
    usr_input = None
    while usr_input not in weekdays:
        usr_input = input('Enter day of week: ')
    
    # display result
    output = work_status(usr_input, weekdays)
    print(output)

our_app()
```
