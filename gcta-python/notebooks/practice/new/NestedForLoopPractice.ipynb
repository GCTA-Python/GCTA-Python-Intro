```python
# Example Problem
# Write out each iteration's i and j values, and any changing variables
# Include changes from any else clause(s)

store = 0
for i in range(2):
    for j in range(3):
        if i == j:
            store += 1
else:
     store += 1

# i = [0, 1]
# j = [0, 1, 2]
    
# iteration 1:
# store starts as 0
# i = 0
# j = 0
# store ends as 1
    
# iteration 2:
# store starts as 1
# i = 0
# j = 1
# store ends as 1
    
# iteration 3:
# store starts as 1
# i = 0
# j = 2
# store ends as 1

# iteration 4:
# store starts as 1
# i = 1
# j = 0
# store ends as 1

# iteration 4:
# store starts as 1
# i = 1
# j = 1
# store ends as 2

# iteration 5:
# store starts as 2
# i = 1
# j = 2
# store ends as 2

# else clase is executed because no break was executed
# else clause adds 1 to store
# store is now 3    

# What is store?
# store is 3
```
